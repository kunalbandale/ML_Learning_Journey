### importing libraries 


```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

### Read CSV file and Display

```python
df = pd.read_csv('lic_data.csv')
```

```python
df
```

### visualize the data using sns

```python
sns.lmplot(x='age',y='premium',data=df) # plot with best fit line
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['age']] , df['premium'])
```

```python
reg.predict([[21]])
```

```python
reg.predict([[50]])
```

### validating using y = mx+c
